```python
Q1 = decode_df['TotalPrice'].quantile(0.25)
Q3 = decode_df['TotalPrice'].quantile(0.75)

IQR = Q3 - Q1
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```python
outliers = decode_df[
    (decode_df['TotalPrice'] < lower_limit) |
    (decode_df['TotalPrice'] > upper_limit)
]

print(outliers)
print("Number of outliers:", len(outliers))
outliers[['Date', 'Product', 'Quantity', 'UnitPrice', 'TotalPrice']]
```

# CUSTOMER SEGMENTATION

```python
# Unsupervised machine learning
# Customer segmentation was performed using unsupervised machine learning (K-Means clustering) 
# to group customers based on purchasing behavior without predefined labels.
```

```python
customer_df = decode_df.groupby('CustomerID').agg(
    Total_Sales=('TotalPrice', 'sum'),
    Total_Quantity=('Quantity', 'sum'),
    Frequency=('OrderID', 'nunique')
).reset_index()
```

```python
customer_df.head()
```

```python
customer_df.shape
```

```python
# 1,189 unique customers
# 11 additional orders belong to customers who purchased more than once. This could have sum it a total of 1200 orders
```

```python
print(decode_df['CustomerID'].nunique())
repeat_customers = decode_df['CustomerID'].value_counts()
print(repeat_customers[repeat_customers > 1])
```

```python
#1,189 unique customers generated 1,200 orders.
# This indicates that some customers returned to make additional purchases, 
# showing evidence of customer retention and repeat purchasing behavior.
```

```python
# Scaling the values using Standardization. This scales from -1 to 1
X = customer_df[['Total_Sales', 'Total_Quantity', 'Frequency']]
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled[:5])
```

```python
# Without scaling:
# Total_Sales would dominate the clustering because it has much larger numbers.
# Quantity and Frequency would have very little influence.
# Scaling ensures all three features contribute fairly to the customer segments.
```

```python
# Finding the best number of clusters(ELBOW METHOD)
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
X
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
customer_df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# SALES VS FREQUENCY
plt.figure()

plt.scatter(customer_df['Total_Sales'],
            customer_df['Frequency'],
            c=customer_df['Cluster'])

plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.title("Customer Segmentation (Sales vs Frequency)")
plt.show()
```

```python
# QUANTTITY VS SALES
plt.figure()

plt.scatter(customer_df['Total_Sales'],
            customer_df['Total_Quantity'],
            c=customer_df['Cluster'])

plt.xlabel("Total Sales")
plt.ylabel("Total Quantity")
plt.title("Customer Segmentation (Sales vs Quantity)")
plt.show()
```

```python
sns.countplot(x='Cluster', data=customer_df)
plt.title("Number of Customers per Segment")
plt.show()
```

```python
cluster_summary = customer_df.groupby('Cluster')['Total_Sales'].mean()
colors = {0: 'red', 1: 'green', 2: 'blue'}
cluster_summary.plot(kind='bar')
plt.title("Average Sales per Cluster")
plt.ylabel("Average Total Sales")
plt.show()
```

```python
customer_df['Segment'] = customer_df['Cluster'].map({
    0: 'Low Value',
    1: 'Medium Value',
    2: 'High Value'
})
```

```python
import matplotlib.pyplot as plt

colors = {0: 'red', 1: 'green', 2: 'blue'}

plt.figure(figsize=(8,5))

for cluster in customer_df['Cluster'].unique():
    data = customer_df[customer_df['Cluster'] == cluster]
    
    plt.scatter(
        data['Total_Sales'],
        data['Total_Quantity'],
        color=colors[cluster],
        label=f'Cluster {cluster}'
    )

plt.xlabel("Total Sales")
plt.ylabel("Total Quantity")
plt.title("Customer Segments")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

colors = {
    'Low Value': '#FF6B6B',
    'Medium Value': '#FFD93D',
    'High Value': '#4D96FF'
}

plt.figure(figsize=(10,6))

for segment in customer_df['Segment'].unique():
    data = customer_df[customer_df['Segment'] == segment]
    
    plt.scatter(
        data['Total_Sales'],
        data['Total_Quantity'],
        label=segment,
        color=colors[segment],
        alpha=0.7,
        edgecolors='black'
    )

# Centroids (important for interpretation)
centroids = customer_df.groupby('Segment')[['Total_Sales','Total_Quantity']].mean()

plt.scatter(
    centroids['Total_Sales'],
    centroids['Total_Quantity'],
    s=300,
    c='black',
    marker='X',
    label='Centroids'
)

plt.title("Customer Segmentation (Business View)", fontsize=14, fontweight='bold')
plt.xlabel("Total Sales")
plt.ylabel("Total Quantity")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)

plt.show()
```
